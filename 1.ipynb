```python
#one page two plots
plt.figure()  #initilizing the page yo make plot
plt.subplot(1,2,1)   #one row two coloumns (1,2) it means that...whose names are one and second...
plt.plot(x , np.sin(x))

plt.subplot(1,2,2)
plt.plot(x , np.cos(x))

plt.show()
```

```python
#one page two plots
plt.figure()  #initilizing the page yo make plot
plt.subplot(2,1,1)   #one row two coloumns (1,2) it means that...whose names are one and second...
plt.plot(x , np.sin(x))

plt.subplot(2,1,2)
plt.plot(x , np.cos(x))

plt.show()
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(x , np.sin(x))

plt.subplot(2,2,2)
plt.plot(x , np.cos(x))

plt.subplot(2,2,3)
plt.plot(x , np.tan(x))

plt.subplot(2,2,4)
plt.plot(x , np.sin(x+1))

plt.show()
```

```python
#piechart >> to understand the percent dist of data
lans = ['C' , 'Python', 'Java']
students = [20,100,10]

plt.pie(students, labels = lans, autopct = "%1.1f%%") #1.1 means two decimal place and f means frequency...
plt.title("Pie chart")
plt.show()
```

```python
#focus on one segment
explode = (0.0, 0.1, 0.1)
lans = ['C' , 'Python', 'Java']
students = [20,100,10]

plt.pie(students, labels = lans, autopct = "%1.1f%%", explode = explode, shadow = True) #1.1 means two decimal place and f means frequency...
plt.title("Pie chart")
plt.show()
```

```python
years = [2021,2018,2022,2023,2017]
category1 = [20,25,30,28,35]
category2 = [10,15,20,18,25]
category3 = [5,8,10,9,15]


plt.stackplot(years, category1,category3,category3, labels = ['category1', 'category2', 'category3'],
              colors = ['skyblue', 'lightgreen', 'orange'])
plt.xlabel("years")
plt.ylabel("values")
plt.title("area plot")
plt.legend()
plt.show()
```
